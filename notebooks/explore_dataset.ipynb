# Explore Combinatorial Optimization Dataset

This notebook loads the unified `all_problems.json` file and produces
basic statistics and visualisations.

```python
import json
from pathlib import Path

DATA_FILE = Path('../data/processed/all_problems.json')

if DATA_FILE.exists():
    with open(DATA_FILE, encoding='utf-8') as fh:
        problems = json.load(fh)
    print(f'Loaded {len(problems)} problems.')
else:
    print('all_problems.json not found. Run the pipeline first:')
    print('  python pipeline/run_collection.py')
    problems = []
```

## Basic Statistics

```python
from collections import Counter

print(f'Total problems: {len(problems)}')

# Problems per source
by_source = Counter(p.get('source_dataset', 'unknown') for p in problems)
print('\nProblems per source:')
for src, count in sorted(by_source.items()):
    print(f'  {src:<35} {count:>5}')

# Problems per category
by_category = Counter(p.get('category', 'unknown') for p in problems)
print('\nProblems per category:')
for cat, count in sorted(by_category.items()):
    print(f'  {cat:<35} {count:>5}')
```

## Sample Problem Entry

```python
if problems:
    import pprint
    pprint.pprint(problems[0], depth=3)
```

## Distribution of Problem Types (LP vs ILP vs MILP)

```python
def classify_problem_type(problem):
    var_types = {
        v.get('type', '')
        for v in problem.get('formulation', {}).get('decision_variables', [])
    }
    if 'binary' in var_types or 'integer' in var_types:
        if 'continuous' in var_types:
            return 'MILP'
        return 'ILP'
    return 'LP'

by_type = Counter(classify_problem_type(p) for p in problems)
print('Problem type distribution:')
for ptype, count in sorted(by_type.items()):
    print(f'  {ptype:<10} {count:>5}')
```

## Visualisation (requires matplotlib)

```python
try:
    import matplotlib.pyplot as plt

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # By source
    axes[0].bar(by_source.keys(), by_source.values())
    axes[0].set_title('Problems by Source')
    axes[0].set_xlabel('Source')
    axes[0].set_ylabel('Count')
    axes[0].tick_params(axis='x', rotation=30)

    # By type
    axes[1].pie(by_type.values(), labels=by_type.keys(), autopct='%1.1f%%')
    axes[1].set_title('Problem Types')

    plt.tight_layout()
    plt.show()
except ImportError:
    print('matplotlib not installed; skipping plot.')
```
